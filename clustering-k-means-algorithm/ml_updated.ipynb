```python
import requests
import json 
import numpy as np
import random
import pandas as pd
import math
import time
import serial 
import astropy.coordinates as astropyc 

from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
from collections import Counter
from numpy import arctan2,random,sin,cos,degrees
from sklearn.metrics import silhouette_score
```

```python
#Lat and Lons for the selected area
start_lat = -0.4081695078335095
start_lon = 36.93586523335967
end_lat = -0.3859893560843097
end_lon = 36.97606395917462

# convert start and end lats and lons to integer
exp = math.pow(10,16)
start_lat = int(start_lat*exp)
start_lon = int(start_lon*exp)
end_lat = int(end_lat*exp)
end_lon = int(end_lon*exp) 

earth_radius = 6.371*math.pow(10,6) 

count = 1

while(True):
    print('Going for process, round {}'.format(count))
   
    #get random co-ordinates
    random_lats = random.sample(range(start_lat, end_lat), 200)
    random_lats = [lat/exp for lat in random_lats] 

    random_lons = random.sample(range(start_lon, end_lon), 200)
    random_lons = [lon/exp for lon in random_lons]
    random_locs = [] 
    z_intilize = 0
    for i in range(len(random_lats)):
        x,y,z= astropyc.spherical_to_cartesian(earth_radius, random_lats[i]*math.pi/180, random_lons[i]*math.pi/180)        
        random_locs.append([x, z])
        z_intilize = z
    print('Printing the first 10 cordinates generated')  
    print(random_locs[:10])   
    print(random_lons[:10])
    print(random_lats[:10])
    loc_compact  = []     
    loc_compact = np.array(random_locs)
    print('Ploting Scatter Plot\n')
    plt.figure(figsize=(12,8))
    plt.scatter(loc_compact[:,0], loc_compact[:,1])
    wcss = []
    print('Determining the optimul number of clusters using the Sihoute Algorithm \n')
    
    for i in range(1, 11):
        kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=len(loc_compact), n_init=10, random_state=0)
        kmeans.fit(loc_compact)
        wcss.append(kmeans.inertia_)
    plt.figure(figsize=(12,8))
    plt.plot(range(1, 11), wcss)
    plt.title('Elbow Method')
    plt.xlabel('Number of clusters')
    plt.ylabel('WCSS')
    plt.show()  
    
    # dissimilarity would not be defined for a single cluster, thus, minimum number of clusters should be 2
    sil = []
    kmax = 10
    print('Determining the optimal number of clusters using Silhoute Algorithm')
    for k in range(2, kmax+1):
      kmeans = KMeans(n_clusters = k).fit(random_locs)
      labels = kmeans.labels_
      sil.append(silhouette_score(random_locs, labels, metric = 'euclidean'))
    print(sil)
    print('Determining the centroid for each cluster\n')
    kmeans = KMeans(n_clusters=5, init='k-means++', max_iter=len(loc_compact), n_init=10, random_state=0)
    pred_y = kmeans.fit_predict(loc_compact)
    plt.figure(figsize=(12,8))
    plt.scatter(loc_compact[:,0], loc_compact[:,1])
    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=len(loc_compact), c='red')
    plt.show()
    centroids = kmeans.cluster_centers_
    labels =list(kmeans.labels_)
    classes = list(set(labels))
    unique = Counter(labels)
    bigger_cluster = max(unique, key=unique.get)
    centroids = centroids[bigger_cluster]
    x = centroids[0]
    y = centroids[1] 
    
    
    #convert back to spherical
    distance, lat, long = astropyc.cartesian_to_spherical(x,y,z)
    lat = str(lat)
    long = str(long)
    lat = lat.replace('rad','')
    long = long.replace('rad','')  
    lat = float(lat)
    long = float(long)
    print('Calculating the bearing of the centroid of the biggest cluster\n')
    a = {'lat': -0.399754, 'long': 36.96256867}
    b = {'lat': lat, 'long': long}
    dL = b['long'] - a['long']
    X = cos(b['lat'])* sin(dL)
    Y = cos(a['lat'])*sin(b['lat']) - sin(a['lat'])*cos(b['lat'])* cos(dL)
    bearing_rad = arctan2(X,Y)
    bearing_degrees = bearing_rad * (180/math.pi)
    bearing_degrees = abs(bearing_degrees)
    start_ta = 'TA{}'.format(random.randint(0,100))
    dest_ta = 'TA{}'.format(random.randint(0,100))
    print('The Bearing of the Centroid of the biggest Cluster is {}'.format(bearing_degrees))
    print('Sending the Bearing of the Biggest Cluster Centroid to Operator for Approval\n')
    url = 'http://odapapi.defrilab.com/api/Odap/addOdapRAngle'
    url_2 = 'https://localhost:44305/api/Odap/addOdapRAngle'
    HEADERS = {
    'Authorization': 'password',
    'Content-Type': 'application/json'
    }
    bearing_degrees = str(bearing_degrees)
    bearing_degrees = bearing_degrees.replace('rad', '')    
    payload = {
            "angle": int(float(bearing_degrees)),
            "dateReceived": "2022-03-09T11:57:15.6469302",
            "isApproved": False,
            "locationFrom": start_ta,
            "locationTo": dest_ta,
            "emailToSendTo": "humphry.shikunzi@outlook.com",
            "id": 0
          }
    response = requests.post(url=url_2 , data=json.dumps(payload), headers=HEADERS, verify=False)
    print(response)
    print('Data Sent Successfully for Approval')
    print('Going into sleep for 10 seconds before redoing the whole process\n')
    count +=1
    time.sleep(60)
```

```python

def polar2cart(r=earth_radius, theta, phi):
    return [
         r * math.sin(theta) * math.cos(phi),
         r * math.sin(theta) * math.sin(phi),
         r * math.cos(theta)
    ]
```
